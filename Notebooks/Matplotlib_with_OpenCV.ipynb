```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
%matplotlib inline
```

```python
img = cv2.imread('C:\\Users\\\80SM01YGIH\\Anish\\CV101\\gradient.png', -1)
_, th1 = cv2.threshold(img, 50, 255, cv2.THRESH_BINARY)
_, th2 = cv2.threshold(img, 200, 255, cv2.THRESH_BINARY_INV)
_, th3 = cv2.threshold(img, 50, 255, cv2.THRESH_TRUNC)
_, th4 = cv2.threshold(img, 50, 255, cv2.THRESH_TOZERO)
_, th5 = cv2.threshold(img, 50, 255, cv2.THRESH_TOZERO_INV)#cv2.imshow('img', img)

titles = ['ORIGINAL', 'BINARY', 'BINARY_INV', 'THRESH_TRUNC', 'THRESH_TOZERO', 'THRESH_TOZERO_INV']
images = [img, th1, th2, th3, th4, th5]

a = 0

fig, axs = plt.subplots(2, 3, figsize = (10, 10))

for i in range(2):
    for j in range(3):
        axs[i][j].imshow(images[a])
        axs[i][j].set_title(titles[a])
        
        a+=1

plt.show()
```
