```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
shapes = cv2.imread('C:\\Users\\80SM01YGIH\\Anish\\CV101\\Color.png', -1)
dupe = shapes.copy()
shapes_gray = cv2.cvtColor(shapes, cv2.COLOR_BGR2GRAY)
_, thresh = cv2.threshold(shapes_gray, 240, 255, cv2.THRESH_BINARY)
contours, _ = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)

for contour in contours:
    approx = cv2.approxPolyDP(contour, 0.01* cv2.arcLength(contour, True), True)
    cv2.drawContours(shapes, [approx], 0, (0, 0, 0), 5)
    x, y = approx.ravel()[0], approx.ravel()[1]
    if len(approx) == 3:
        cv2.putText(shapes, 'Triangle', (x, y), cv2.FONT_HERSHEY_COMPLEX, 0.5, (0, 0, 0))
    elif len(approx) == 4:
        x, y, w, h = cv2.boundingRect(approx)
        aspect_ratio = float(w)/h
        print(aspect_ratio)
        if (aspect_ratio >= 0.95 and aspect_ratio <= 1.05):
            cv2.putText(shapes, 'Square', (x, y), cv2.FONT_HERSHEY_COMPLEX, 0.5, (0, 0, 0))
        else:
            cv2.putText(shapes, 'Rectangle', (x, y), cv2.FONT_HERSHEY_COMPLEX, 0.5, (0, 0, 0))
    elif len(approx) == 5:
        cv2.putText(shapes, 'Pentagon', (x, y), cv2.FONT_HERSHEY_COMPLEX, 0.5, (0, 0, 0))
    elif len(approx) == 10:
        cv2.putText(shapes, 'Star', (x, y), cv2.FONT_HERSHEY_COMPLEX, 0.5, (0, 0, 0))
    else:
        cv2.putText(shapes, 'Circle', (x, y), cv2.FONT_HERSHEY_COMPLEX, 0.5, (0, 0, 0))

cv2.imshow('original image', shapes)
cv2.imshow('image', dupe)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
