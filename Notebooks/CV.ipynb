```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#img = cv2.imread(cv2.samples.findFile('messi5.jpg'))
img = cv2.imread('C:\\Users\\80SM01YGIH\\Anish\\CV101\\messi5 - Copy.jpg', -1)
plt.imshow(img)
```

```python
plt.imshow(img)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsl = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
img_types = [img_gray, img_rgb, img_hsl]

fig, axs = plt.subplots(1, 4, figsize = (20, 20))
for i in range (0,3):
    axs[i].imshow(img_types[i], cmap = 'gray')
axs[3].imshow(img)
```

```python
cv2.imwrite('messy_copy.jpg', img_hsl)
```

```python
new_img = img.copy()
cv2.imshow('Messi copy', new_img)
k = cv2.waitKey(0)
if k == 27:
    cv2.destroyAllWindows()
elif k == ord('s'):
    cv2.imwrite('new_image', new_img)
    cv2.destroyAllWindows()
```
