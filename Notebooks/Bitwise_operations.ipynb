```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = np.zeros((250, 500, 3), np.uint8)
img1 = cv2.rectangle(img1, (200, 0), (300, 100), (255, 255, 255), -1)
img2 = np.zeros((250, 500, 3), np.uint8)
pts = np.array([[250, 0], [250, 250], [500, 250], [500, 0]], np.int32)
img2 = cv2.fillPoly(img2, [pts], (255, 255, 255))

bitwiseA = cv2.bitwise_and(img1, img2)
bitwiseO = cv2.bitwise_or(img1, img2)
bitwiseX = cv2.bitwise_xor(img1, img2)
bitwiseN = cv2.bitwise_not(img1)

fig, ax = plt.subplots(3, 2, figsize = (10, 10))
ax[0][0].imshow(img1)
ax[0][1].imshow(img2)
ax[1][0].imshow(bitwiseA)
ax[1][1].imshow(bitwiseO)
ax[2][0].imshow(bitwiseX)
ax[2][1].imshow(bitwiseN)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
