```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('C:\\Users\\80SM01YGIH\\Anish\\CV101\\lena.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

kernel = np.ones((5, 5), np.float32)/25
dst = cv2.filter2D(img, -1, kernel)
blur = cv2.blur(img, (5, 5))
gaussianblur = cv2.GaussianBlur(img, (5, 5), 0)
median = cv2.medianBlur(img, 5) #Kernel size should not be 1
bilateralfilter = cv2.bilateralFilter(img, 9, 75, 75)#The keep the edges sharper even if the image is blurred,
                                                     #Compare the 2-D convolution you can see that the edge preservation are more effective

x = 6

titles = ['Original', '2D - Convolution', 'Blur', 'GaussianBlur', 'MedianBlur', 'BilateralFilter']
images = [img, dst, blur, gaussianblur, median, bilateralfilter]
```

```python
plt.figure(figsize = (10, 7))
for i in range(x):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], cmap = 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
plt.show()
```
