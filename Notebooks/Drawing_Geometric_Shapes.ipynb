```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, axs = plt.subplots(3, 3, figsize = (20, 10))
#img = cv2.imread('C:\\Users\\80SM01YGIH\\Anish\\CV101\\messi5 - Copy.jpg', -1)
#img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img = np.zeros([512, 512, 3], np.uint8)
#Draw a straight line
straight_line = cv2.line(img, (255, 255), (255, 100), (0, 0, 255), 10)
axs[0][1].imshow(straight_line)

#Draw an arrowed line
arrowed_line = cv2.arrowedLine(img, (0, 255), (200, 255), (0, 255, 0), 10)
axs[0][2].imshow(arrowed_line)

#Draw a rectangle
draw_rectangle = cv2.rectangle(img, (0, 0), (200, 50), (0, 0, 255), -1)
axs[1][0].imshow(draw_rectangle)

#Draw a circle
draw_circle = cv2.circle(img, (100, 100), 100, (0, 255, 0), -1)
axs[1][1].imshow(draw_circle)

#Adding a text
font = cv2.FONT_HERSHEY_SIMPLEX
add_text = cv2.putText(img, 'Messi', (0, 100), font, 2, (0, 0, 255), 5, cv2.LINE_AA)
axs[1][2].imshow(add_text)

#Drawing a polynomial
pts = [[1,100], [100,200], [200,150]]
polygon = cv2.polylines(img, [pts], True, (200, 0, 0))
axs[2][0].imshow(polygon)

axs[0][0].imshow(img, cmap = 'gray')
```
