```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('C:\\Users\\80SM01YGIH\\Anish\\CV101\\messi5 - Copy.jpg', 0)

laplacian = cv2.Laplacian(img, cv2.CV_64F)
laplacian = np.uint8(np.absolute(laplacian))

#Sobelx and sobely are also known as sobel gradient
sobelX = cv2.Sobel(img, cv2.CV_64F, 1, 0) #1 is for SobelX
sobelX = np.uint8(np.absolute(sobelX))
sobelY = cv2.Sobel(img, cv2.CV_64F, 0, 1) #1 is for SobelY
sobelY = np.uint8(np.absolute(sobelY))

sobel_gradient = cv2.bitwise_or(sobelX, sobelY)

edges = cv2.Canny(img, 100, 200)

x = 6

titles = ['Original', 'Laplacian', 'sobelX', 'sobelY', 'Sobel_gradient', 'Canny edge detection']
images = [img, laplacian, sobelX, sobelY, sobel_gradient, edges]
```

```python
plt.figure(figsize = (20, 10))
for i in range(x):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
